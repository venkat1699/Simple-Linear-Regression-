```python
import warnings
warnings.filterwarnings('ignore')
import statsmodels.formula.api as smf
model = smf.ols('Experience ~ Salary',data=dt).fit()
```

```python
model.summary()
```

```python
model.params
```

```python
model.rsquared, model.rsquared_adj
```

```python
model.pvalues, model.tvalues
```

```python
pred = model.predict(dt.Salary)
pred
```

```python
rmse_lin= np.sqrt(np.mean((np.array(dt['Experience'])-np.array(pred))**2))
rmse_lin
```

```python
plt.scatter(x=dt['Salary'],y = dt['Experience'],color ='red')
plt.plot(dt['Salary'],pred,color='black')
plt.xlabel('Salary')
plt.ylabel('Experience')
plt.show()
```

```python
model2 = smf.ols('Experience ~ np.log(Salary)',data=dt).fit()
```

```python
model2.summary()
```

```python
pred2 = model2.predict(pd.DataFrame(dt['Salary']))
pred2
```

```python
rmse_log= np.sqrt(np.mean((np.array(dt['Experience'])-np.array(pred2))**2))
rmse_log
```

```python
plt.scatter(x=dt['Salary'],y = dt['Experience'],color ='red')
plt.plot(dt['Salary'],pred2,color='black')
plt.xlabel('Salary')
plt.ylabel('Experience')
plt.show()
```

```python
model3 = smf.ols('np.log(Experience) ~ Salary',data=dt).fit()
model3.summary()
```

```python
pred_log = model3.predict(dt['Salary'])
pred_log
```

```python
pred3=np.exp(pred_log)
pred3
```

```python
rmse_exp= np.sqrt(np.mean((np.array(dt['Experience'])-np.array(pred3))**2))
rmse_exp
```

```python
plt.scatter(x=dt['Salary'],y = dt['Experience'],color ='red')
plt.plot(dt['Salary'],pred3,color='black')
plt.xlabel('Salary')
plt.ylabel('Experience')
plt.show()
```

```python
dt['Salary_sqr'] = dt.Salary*dt.Salary
dt
```

```python
model4 = smf.ols('np.log(Experience)~ Salary+Salary_sqr',data=dt).fit()
model4.summary()
```

```python
pred_exp = model4.predict(dt)
pred4=np.exp(pred_exp)
pred4
```

```python
rmse_quad = np.sqrt(np.mean((np.array(dt['Experience'])-np.array(pred4))**2))
rmse_quad
```

```python
plt.scatter(dt.Salary,dt.Experience, color = 'green')
plt.plot(dt.Salary,pred4, 'r')
```

```python
model4.resid_pearson
```
